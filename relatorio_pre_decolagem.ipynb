### **2. Planejar algoritmo**

Regras de segurança definidas:
- Temperatura interna: 18°C a 26°C
- Temperatura externa: 15°C a 35°C
- Integridade estrutural: deve ser sempre 1
- Nível de energia: mínimo 95%
- Pressão dos tanques: 240 a 255 bar
- Status dos módulos críticos: deve ser OK

Pseudocódigo:
```
INICIO

CARREGAR CSV EM dados_telemetria

falha_critica = FALSO
contador_falhas = 0

PARA CADA registro EM dados_telemetria FACA

  temp_interna = registro.temperatura_interna_c
  temp_externa = registro.temperatura_externa_c
  integridade  = registro.integridade_estrutural
  energia      = registro.nivel_energia_percentual
  pressao      = registro.pressao_tanques_bar
  status       = registro.status_modulos_criticos

  SE integridade == 0
    falha_critica = VERDADEIRO

  SE status != 'OK'
    falha_critica = VERDADEIRO

  SE temp_interna < 18 OU temp_interna > 26
    contador_falhas = contador_falhas + 1

  SE temp_externa < 15 OU temp_externa > 35
    contador_falhas = contador_falhas + 1

  SE energia < 95
    contador_falhas = contador_falhas + 1

  SE pressao < 240 OU pressao > 255
    contador_falhas ← contador_falhas + 1

FIMPARA

SE falha_critica == VERDADEIRO
    IMPRIMIR "DECOLAGEM ABORTADA"
SENAO SE contador_falhas > 3
    IMPRIMIR "DECOLAGEM ABORTADA"
SENAO
    IMPRIMIR "PRONTO PARA DECOLAR"

FIM
```

### **3. Desenvolver script em Python**


````python
import pandas
print('Pandas importado com sucesso')

from google.colab import files
print('Por favor faça upload do arquivo contendo a telemetria')
uploaded = files.upload()

nome_arquivo = list(uploaded.keys())[0]
df = pandas.read_csv(nome_arquivo)
print('Conteúdo do arquivo:')

falha_critica   = False
contador_falhas = 0

for index, row in df.iterrows():

    temp_interna = row['temperatura_interna_c']
    temp_externa = row['temperatura_externa_c']
    integridade  = row['integridade_estrutural']
    energia      = row['nivel_energia_percentual']
    pressao      = row['pressao_tanques_bar']
    status       = row['status_modulos_criticos']

    if integridade == 0:
        falha_critica = True
        break

    if status == "ERRO":
        falha_critica = True
        break

    if temp_interna < 18 or temp_interna > 26:
        contador_falhas += 1

    if temp_externa < 15 or temp_externa > 35:
        contador_falhas += 1

    if energia < 95:
        contador_falhas += 1

    if pressao < 240 or pressao > 255:
        contador_falhas += 1

if falha_critica:
    print("FALHA CRITICA DETECTADA, DECOLAGEM ABORTADA")
elif contador_falhas > 3:
    print("DECOLAGEM ABORTADA")
else:
    print("PRONTO PARA DECOLAR")
````

### **4. Análise energética**

````python
capacidade_total_kwh = 5000
carga_atual = df['nivel_energia_percentual'].iloc[-1]
consumo_decolagem_kwh = 2000
perdas_kwh = 200

energia_disponivel = capacidade_total_kwh * (carga_atual / 100)
autonomia = (energia_disponivel - perdas_kwh) / consumo_decolagem_kwh

print(f"Energia disponível: {carga_atual:.2f}%, {energia_disponivel:.2f} kWh")
print(f"Autonomia estimada para decolagem: {autonomia:.2f} ciclos de decolagem")

if autonomia >= 1:
    print("Energia suficiente para decolagem")
else:
    print("Energia INSUFICIENTE para decolagem")
````
